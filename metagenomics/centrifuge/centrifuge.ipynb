```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import taxopy
```

```python
df = pd.read_csv('data/centrifuge_report/report.N23.tsv', sep='\t')
df = df.sort_values(by="abundance", ascending=False)
```

```python
taxdb = taxopy.TaxDb(nodes_dmp="../taxdb/nodes.dmp", names_dmp="../taxdb/names.dmp")
kingdoms = []
for row in range(len(df)):
    taxid = df.iloc[row]['taxID']
    key = "kingdom"
    dict = taxopy.Taxon(taxid, taxdb).rank_name_dictionary
    if key in dict:
        kingdoms.append(taxopy.Taxon(taxid, taxdb).rank_name_dictionary["kingdom"])
    else:
        kingdoms.append("NA")

df['kingdom'] = kingdoms
```
