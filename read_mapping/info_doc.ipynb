Competetive read mapping for samples:

```python
### Displaying Figures

import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

figures_path = './figures/comp3/all_cov/samples'
figures = [f for f in os.listdir(figures_path) if f.endswith('.png') or f.endswith('.jpg')]

for figure in figures:
    img = mpimg.imread(os.path.join(figures_path, figure))
    plt.figure()
    plt.imshow(img)
    plt.axis('off')
    plt.title(figure)
    plt.show()
```
