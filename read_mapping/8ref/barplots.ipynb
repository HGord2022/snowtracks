```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("coverages_comp.csv")  # your table saved as data.csv

for _, row in df.iterrows():
    sample = row["sample"]
    values = row.drop("sample")

    plt.figure()
    values.plot(kind="bar")
    plt.title(sample)
    plt.ylabel("Value")
    plt.tight_layout()
    plt.savefig(f"{sample}_barplot.png", dpi=300)
    plt.close()
```
